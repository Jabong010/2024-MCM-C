```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件并创建DataFrame
data = pd.read_csv('data.csv')

# 统计不同性别的数量
gender_count = data['Sex'].value_counts()

# 绘制扇形图
plt.figure(figsize=(6, 6))
plt.pie(gender_count, autopct='%1.1f%%', startangle=0, wedgeprops=dict(edgecolor='black'))
plt.title('Gender')
plt.axis('equal')  # 保证是一个圆形

# 在图表上添加 "male" 和 "female" 文字标签
plt.text(-0.9, 0.8, "Male", color='black', fontsize=12, ha='center', va='center')
plt.text(0.9, -0.8, "Female", color='black', fontsize=12, ha='center', va='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# [省略读取CSV文件和创建DataFrame的代码]
# 填充缺失值为 'No Medal'
data['Medal'].fillna('No Medal', inplace=True)

# 根据性别筛选数据
male_data = data[data['Sex'] == 'M']
female_data = data[data['Sex'] == 'F']

# 统计男性和女性获得不同奖牌的数量
male_medals = male_data['Medal'].value_counts()
female_medals = female_data['Medal'].value_counts()

# 获取奖牌类型
medal_types = data['Medal'].unique()

# 创建画布和子图
fig, ax = plt.subplots(figsize=(10, 6))

# 绘制男性运动员奖牌分布
plt.bar(male_medals.index + '_M', male_medals.values, color=['#1f77b4', 'orange', 'green', 'red'], label='Male', width=1)

# 绘制女性运动员奖牌分布
plt.bar(female_medals.index + '_F', female_medals.values, color=['#1f77b4', 'orange', 'green', 'red'], label='Female', width=1)

# 设置图例
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in ['#1f77b4', 'orange', 'green', 'red']]
labels = medal_types
plt.legend(handles, labels, title='Medal Types', bbox_to_anchor=(0.8, 1), loc='upper left')

# 添加第二个x轴
ax2 = ax.twiny()
ax2.set_xticks([])
ax2.set_xlabel('Sex')

# 添加第一个x轴标签
ax.set_xticks([1.5, 5.5])
ax.set_xticklabels(['M', 'F'])
ax.set_xlabel('sex')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# 读取CSV文件
data = pd.read_csv('data.csv')

# 获取Games列第一次出现的顺序
games_order = data['Games'].drop_duplicates().reset_index(drop=True)

# 统计不同年份和赛季的奥运会数量
seasons_count = data.groupby(['Year', 'Season', 'Games']).size().reset_index(name='Count')

# 根据Games列的第一次出现顺序排序
seasons_count['Games'] = pd.Categorical(seasons_count['Games'], categories=games_order, ordered=True)
seasons_count = seasons_count.sort_values(by=['Games'])

# Create a custom gradient color from dark green to light green
color_map = mcolors.LinearSegmentedColormap.from_list(
    'custom_green', ['#006400', '#FFFACD'], N=len(seasons_count)
)

# 创建柱状图，使用自定义渐变色
plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(seasons_count)), seasons_count['Count'], color=color_map(np.arange(len(seasons_count))))
plt.xlabel('Games')  # Only display 'Games' on the x-axis
plt.ylabel('Count')

# Set x-axis ticks and labels
plt.xticks(range(len(seasons_count)), seasons_count['Games'], rotation=90)

# Add a colorbar for reference
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=color_map))
cbar.set_label('Color Gradient')

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns

# [省略读取CSV文件和创建DataFrame的代码]
colors = sns.color_palette('viridis', len(seasons_count))[::-1]

# Darken the colors
darkened_colors = [(max(0, c[0] - 0.18), max(0, c[1] - 0.18), max(0, c[2] - 0.15)) for c in colors]

# 统计不同年份的奥运会数量
year_count = data['Year'].value_counts().sort_index(ascending=False)

# 创建横向柱状图，调整图尺寸
plt.figure(figsize=(12, 6))
year_count.plot(kind='barh', color=darkened_colors)  # Use the darkened colors
plt.title('Number of Olympic Games by Year')
plt.xlabel('Count')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

```python
# [省略读取CSV文件和创建DataFrame的代码]

# 选择所需的列：年龄、身高和奖牌种类
selected_data = data[['Age', 'Height', 'Medal']]

# 填充缺失值为 'No Medal'
selected_data['Medal'].fillna('No Medal', inplace=True)

# 去除缺失值
selected_data.dropna(subset=['Age', 'Height'], inplace=True)

# 修正颜色列表的长度，确保与奖牌的唯一值数量相匹配
unique_medals = selected_data['Medal'].unique()
colors = ['#1f77b4', 'orange', 'green', 'red', 'gray'][:len(unique_medals)]

# 创建散点图
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Height', hue='Medal', data=selected_data, palette=colors)
plt.title('Scatter Plot of Age vs Height by Medal Type')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend(title='Medal')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# [省略读取CSV文件和创建DataFrame的代码]

# 选择所需的列：身高、体重和奖牌种类
selected_data = data[['Height', 'Weight', 'Medal']]

# 填充缺失值为 'No Medal'
selected_data['Medal'].fillna('No Medal', inplace=True)

# 去除缺失值
selected_data.dropna(subset=['Height', 'Weight'], inplace=True)

# 修正颜色列表的长度，确保与奖牌的唯一值数量相匹配
unique_medals = selected_data['Medal'].unique()
colors = ['#1f77b4', 'orange', 'green', 'red', 'gray'][:len(unique_medals)]

# 创建散点图
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Height', y='Weight', hue='Medal', data=selected_data, palette=colors)
plt.title('Scatter Plot of Height vs Weight by Medal Type')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(title='Medal')
plt.tight_layout()
plt.show()
```
